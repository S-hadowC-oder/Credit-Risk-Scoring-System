```python
#Checking memory usages
memory_use = df.memory_usage(deep= True).sum() / 1024**2
print(f"\nMeomry usages:{memory_use:.2f}MB")
```

### The :.2f inside the print statement is a string formatting command in Python (used inside f-strings).

Here is exactly what it does:

 : tells Python, "I am about to specify how this variable should be formatted."
 .2 specifies the precision. It means, "Round this number to exactly 2 decimal places."
 f stands for float (a number with a decimal point). It tells Python to display the variable as a standard decimal number, rather than scientific notation.


### Why do we use it in this context?
The result of the memory calculation (df.memory_usage(deep=True).sum() / 1024**2) usually produces a very long floating-point number, something like 948.3847291048 MB.

Reporting a number like that is visually messy and distracting. By adding :.2f, we force Python to round and display it cleanly as 948.38 MB.

Examples to show the difference:

python
>value = 123.456789
>print(f"Normal: {value}")      # Outputs: Normal: 123.456789
>print(f"With .2f: {value:.2f}")  # Outputs: With .2f: 123.46
>print(f"With .0f: {value:.0f}")  # Outputs: With .0f: 123 (Rounds to whole number)
>In production and reporting (like Jupyter Notebooks meant for stakeholders), presenting clean, formatted numbers is always considered best practice!
